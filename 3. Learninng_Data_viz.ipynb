## Two options for subplots

```python
# subplot option 1
fig,((ax1, ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                                         ncols = 2,
                                         figsize =(10,5))

#plot tto each diffeent axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(cars.keys(), height = cars.values());
ax4.hist(x);
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows =2,
                      ncols = 2,
                      figsize = (10,5))
    
    
#plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(cars.keys(), height = cars.values());
ax[1,1].hist(x);
```

## Which plot is ideal ? (pyplot vs matplolib OO method)
when plottign somethng qualiclky is okay to use the pyplot OO method
but when plotting more advanced use the OO object

```python
heart_disease
```

```python
#using bolean index to select patiaent details that is above 50
over_50 = heart_disease[heart_disease["age"] >50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind = "scatter",
            x = "age",
           y = 'chol',
            c= 'target');
```

```python
# Recreating whats above uisng O method
#here the x axisis well lbale

fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind ='scatter',
            x ='age',
            y = 'chol',
            c= 'target',
            ax=ax)
#ax.set_xlim([45,100]) #creting axis limit
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize =(10,6))

# Plot the data
scatter = ax.scatter(x =over_50['age'],
                    y = over_50['chol'],
                    c= over_50['target'])

# Customise your chart
ax.set (title = "Heart disas and Cholesterol Levls",
       xlabel = "Age",
       ylabel = "cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Add horizontal line
ax.axhline(over_50['chol'].mean(),linestyle ="--")
```

```python
#over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                           ncols = 1,
                           figsize =(10,15),
                              sharex = True) # it allow both fig to use same x value

# Add data to ax0

scatter = ax0.scatter(x = over_50["age"],
                 y = over_50['chol'],
                 c = over_50["target"])

# customise daata
ax0.set(title = "Scatter graph of Age vs Chol",
      xlabel = "Age",
      ylabel = "Cholesterol")

#add legend
ax0.legend(*scatter.legend_elements(), title = 'Target')

# add horinzontal line
ax0.axhline(over_50["chol"].mean(), linestyle = "--")

scatter_2 = ax1.scatter( x = over_50["age"],
                                    y = over_50["thalach"],
                       c = over_50["target"])

ax1.set (title = "Scatter graph of Age vs thalach",
        xlabel = "Age",
        ylabel= "Thalach")

ax1.legend(*scatter_2.legend_elements(),title ="Target")


ax1.axhline(over_50['thalach'].mean(), linestyle = "--")


#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight ='bold')
```

```python
# Customising the y and x indentation
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                           ncols = 1,
                           figsize =(10,15),
                              sharex = True) # it allow both fig to use same x value

# Add data to ax0

scatter = ax0.scatter(x = over_50["age"],
                 y = over_50['chol'],
                 c = over_50["target"],
                     cmap="winter")

# customise data
ax0.set(title = "Scatter graph of Age vs Chol",
      xlabel = "Age",
      ylabel = "Cholesterol")

#add legend
ax0.legend(*scatter.legend_elements(), title = 'Target')

# add horinzontal line
ax0.axhline(over_50["chol"].mean(), linestyle = "--")

#customize axes 0
ax0.set_xlim(50,80)

#plot second plot
scatter_2 = ax1.scatter( x = over_50["age"],
                                    y = over_50["thalach"],
                       c = over_50["target"],
                       cmap="winter")

ax1.set (title = "Scatter graph of Age vs thalach",
        xlabel = "Age",
        ylabel= "Thalach")

ax1.legend(*scatter_2.legend_elements(),title ="Target")


ax1.axhline(over_50['thalach'].mean(), linestyle = "--")

#customize axes 1
ax1.set_xlim(50,80)
ax1.set_ylim(60,200)

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight ='bold')
```
